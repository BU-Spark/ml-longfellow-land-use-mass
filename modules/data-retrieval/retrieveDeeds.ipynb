```python
import os
from google.colab import drive
drive.mount('/content/drive')
```

```python
def retrieve_deeds(dir, max):
  deeds = []
  for name in os.listdir(dir):
    if len(deeds) >= max:
      break
    if name.endswith('TIF'):
      deeds.append(dir + '/' + name)
    else:
      print(dir + '/' + name)
      new = retrieve_deeds(dir + '/' + name, max-len(deeds))
      deeds.extend(new)
  return deeds
```

```python
def retrieve_foldernames(dir):
  names = []
  for name in os.listdir(dir):
    if not name.endswith('TIF'):
      names.append(dir + '/' + name)
      new = retrieve_foldernames(dir + '/' + name)
      names.extend(new)
  return names
```

```python
# total number of TIFs is 42200
print(retrieve_deeds('/content/drive/MyDrive/Colab Notebooks/Data for Fall 2024 Project', 100))
# can retrieve all folder names, if want data from specific folders
print(retrieve_foldernames('/content/drive/MyDrive/Colab Notebooks/Data for Fall 2024 Project'))
```
